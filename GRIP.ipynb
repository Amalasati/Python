IMPORTING LIBRARIEA

```python
# Importing all required libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
data.head(9)
```

```python
# Checking Dimensions of Data
print(data)
print("Dimensions:")
print(data.shape)
```

```python
# taking out statistical summary of data
print("statistics Summary-")
print(data.describe())
```

```python
#checking missing values
data.isnull().values.any()
```

```python
#plotting scatter plot
data.plot(kind="scatter", x="Hours", y="Scores",color="r")
plt.show()
```

```python
data.corr(method="pearson")
```

```python
Hours=data["Hours"]
Score=data["Scores"]
sns.displot(Hours)
sns.displot(Score)
```

# **Linear Regression**

```python
#Splitting data in x and y
x=data.drop("Scores",axis=1)
y=data["Scores"]
```

```python
#Now splitting data to train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
a=lr.coef_
c=lr.intercept_
line=a*x+c
plt.scatter(x,y)
plt.plot(x,line)
plt.show()
```

```python
y_pred=lr.predict(x_test)
df=pd.DataFrame({"Actual":y_test,"Predict":y_pred})
print(df)
print("Accuracy:",lr.score(x,y)*100)
```

What would be the predicted score if a student studies 9 hours/day

```python
h=9
b=lr.predict([[h]])
print(f"If a student studied for {h} hours/day will score{b}% in exam.")
```

# **Model Evaluation**

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("mean squared error:",mean_squared_error(y_test,y_pred))
print("mean absolute error:",mean_absolute_error(y_test,y_pred))
print("R2 score:",r2_score(y_test,y_pred))
```
